<h1>Graphiques de comparaisons des temps d'exécution et de la longueur des solutions trouvées pour les planners HSP et Random Walk</h1>

Lors de la collecte de données, le temps de recherche des planners a été limité à 15 min (c'est-à-dire 900 000 millisecondes), car certains problèmes prennent un temps conséquent à s'exécuter.

Dans les graphiques ci-dessous, lorsqu'un planner n'a pas trouvé de solution ou qu'il est arrivé au bout du temps d'exécution qui lui a été imparti, son temps d'exécution est mis à 900 000 millisecondes et la longueur du plan solution est mise à 0.

```python
import csv
import matplotlib.pyplot as plt

# Fonction pour charger les données depuis le fichier texte
def load_dataL(filename):
    data = []
    with open(filename, mode='r', newline='', encoding='utf-8') as file:
        reader = csv.reader(file, delimiter=';')
        next(reader)  # Ignorer l'en-tête
        for row in reader:
            problem = row[0]
            # Remplacer les valeurs manquantes par 0, si nécessaire
            hsplength = int(row[1]) if row[1] else 0
            rwlength = int(row[2]) if row[2] else 0
            data.append({'probleme': problem, 'HSPlength': hsplength, 'SATlength': rwlength})
    return data

def load_dataT(filename):
    data = []
    with open(filename, mode='r', newline='', encoding='utf-8') as file:
        reader = csv.reader(file, delimiter=';')
        next(reader)  # Ignorer l'en-tête
        for row in reader:
            problem = row[0]
            # Remplacer les valeurs manquantes par 0, si nécessaire
            hsptemps = int(row[1]) if row[1] else 600000
            sattemps = int(row[2]) if row[2] else 600000
            data.append({'probleme': problem, 'HSPtemps': hsptemps, 'SATtemps': sattemps})
    return data

# Fonction pour créer le graphique
def plot_dataL(data,jeu):
    # Extraire les noms des problèmes, les longueurs HSP et SAT
    problems = [entry['probleme'] for entry in data]
    hsplengths = [entry['HSPlength'] for entry in data]
    rwlengths = [entry['SATlength'] for entry in data]

    # Création du graphique
    plt.figure(figsize=(10, 6))

    # Affichage des données
    plt.plot(problems, hsplengths, label='HSP', marker='o', color='b', linestyle='-', markersize=5)
    plt.plot(problems, rwlengths, label='SAT', marker='s', color='r', linestyle='-', markersize=5)

    # Ajout des labels et titre
    plt.xlabel('Problème')
    plt.ylabel('Longueur du plan solution')
    plt.title('Comparaison des Longueurs HSPlength et SATlength selon les Problèmes pour le jeu '+jeu)
    plt.xticks(rotation=45, ha='right')  # Rotation des noms de problèmes pour plus de lisibilité
    plt.legend()

    # Affichage du graphique
    plt.tight_layout()  # Ajuste les éléments du graphique pour qu'ils s'affichent correctement
    plt.show()

def plot_dataT(data,jeu):
    # Extraire les noms des problèmes, les temps HSP et SAT
    problems = [entry['probleme'] for entry in data]
    hsptemps = [entry['HSPtemps'] for entry in data]
    sattemps = [entry['SATtemps'] for entry in data]

    # Création du graphique
    plt.figure(figsize=(10, 6))

    # Affichage des données
    plt.plot(problems, hsptemps, label='HSP', marker='o', color='b', linestyle='-', markersize=5)
    plt.plot(problems, sattemps, label='SAT', marker='s', color='r', linestyle='-', markersize=5)

    # Ajout des labels et titre
    plt.xlabel('Problème')
    plt.ylabel('durées en millisecondes')
    plt.title('Comparaison des durées HSPtemps et SATtemps selon les Problèmes pour le jeu '+jeu)
    plt.xticks(rotation=45, ha='right')  # Rotation des noms de problèmes pour plus de lisibilité
    plt.legend()

    # Affichage du graphique
    plt.tight_layout()  # Ajuste les éléments du graphique pour qu'ils s'affichent correctement
    plt.show()


```

```python
filename = 'blocksLength.txt'  # Remplacer par le chemin réel du fichier
data = load_dataL(filename)
plot_dataL(data,"blockworld")

filename = 'blocksTime.txt'  # Remplacer par le chemin réel du fichier
data = load_dataT(filename)
plot_dataT(data,"blockworld")
```
